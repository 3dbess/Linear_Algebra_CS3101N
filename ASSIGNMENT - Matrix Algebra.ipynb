1. Without the use of any python libraries or modules, develop a function that can perform matrix addition, given two numpy matrices.

```python
def matrix_addition(matrix1, matrix2):
    # Check if the matrices have the same dimensions
    if len(matrix1) != len(matrix2) or len(matrix1[0]) != len(matrix2[0]):
        raise ValueError("Matrices must have the same dimensions for addition")

    # Initialize an empty result matrix with the same dimensions
    result = [[0] * len(matrix1[0]) for _ in range(len(matrix1))]

    # Perform element-wise addition
    for i in range(len(matrix1)):
        for j in range(len(matrix1[0])):
            result[i][j] = matrix1[i][j] + matrix2[i][j]

    return result

# Example matrices for testing
matrix_a = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
matrix_b = [[9, 8, 7], [6, 5, 4], [3, 2, 1]]

# Test matrix_addition function
result_matrix = matrix_addition(matrix_a, matrix_b)

# Display the result
print("Matrix A:")
for row in matrix_a:
    print(row)

print("\nMatrix B:")
for row in matrix_b:
    print(row)

print("\nResult Matrix:")
for row in result_matrix:
    print(row)
```

2. Without the use of any python libraries or modules, develop a function that can multiply two numpy matrices.

```python
def matrix_multiplication(matrix1, matrix2):
    # Check if the matrices can be multiplied
    if len(matrix1[0]) != len(matrix2):
        raise ValueError("Number of columns in the first matrix must be equal to the number of rows in the second matrix")

    # Initialize an empty result matrix with dimensions (rows of matrix1) x (columns of matrix2)
    result = [[0] * len(matrix2[0]) for _ in range(len(matrix1))]

    # Perform matrix multiplication
    for i in range(len(matrix1)):
        for j in range(len(matrix2[0])):
            for k in range(len(matrix2)):
                result[i][j] += matrix1[i][k] * matrix2[k][j]

    return result

# Example matrices for testing
matrix_a = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
matrix_b = [[9, 8, 7], [6, 5, 4], [3, 2, 1]]

# Test matrix_multiplication function
result_matrix = matrix_multiplication(matrix_a, matrix_b)

# Display the result
print("Matrix A:")
for row in matrix_a:
    print(row)

print("\nMatrix B:")
for row in matrix_b:
    print(row)

print("\nResult Matrix:")
for row in result_matrix:
    print(row)
```

4. Without using a python library or modules develop a function that can extract the diagonal of a numpy matrix.

```python
def extract_diagonal(matrix):
    # Check if the matrix is square
    if len(matrix) != len(matrix[0]):
        raise ValueError("Input matrix must be square for diagonal extraction")

    # Extract the diagonal elements
    diagonal = [matrix[i][i] for i in range(len(matrix))]
    return diagonal

# Example matrix for testing
example_matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]

# Test extract_diagonal function
diagonal_elements = extract_diagonal(example_matrix)
print("Diagonal Elements:")
print(diagonal_elements)
```

5. Without using a python library or modules develop a function that can find a trace of a numpy matrix.

```python
def matrix_trace(matrix):
    # Check if the matrix is square
    if len(matrix) != len(matrix[0]):
        raise ValueError("Input matrix must be square for trace calculation")

    # Calculate the trace (sum of diagonal elements)
    trace = sum(matrix[i][i] for i in range(len(matrix)))
    return trace

# Example matrix for testing
example_matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]

# Test matrix_trace function
trace_value = matrix_trace(example_matrix)
print("Trace of the Matrix:", trace_value)
```
